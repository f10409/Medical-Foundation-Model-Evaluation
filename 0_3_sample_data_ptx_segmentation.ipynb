```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./inputs/SIIM_ACR_input.csv')
df
```

```python
df_p = df[df.Pneumothorax==1]
df_p
```

```python
# Shuffle the dataframe
df_balanced = df_p.sample(frac=1, random_state=5656).reset_index(drop=True)

# Calculate total samples and samples per fold
n_total = len(df_balanced)
n_per_fold = n_total // 5
remainder = n_total % 5

# Create fold assignments
fold_assignments = []
for fold in range(5):
    # Add one extra sample to first 'remainder' folds to handle uneven division
    fold_size = n_per_fold + (1 if fold < remainder else 0)
    fold_assignments.extend([fold] * fold_size)

# Assign fold numbers to the dataframe
df_balanced['Fold'] = fold_assignments

# Generate 5 files for cross-validation
for fold in range(5):
    # Create a copy of the dataframe
    df_fold = df_balanced.copy()
    
    # Create Split column based on current fold
    # Current fold becomes 'Test'
    # For remaining data, use similar proportions as your original code (75% train, 25% valid)
    def assign_split(fold_num):
        if fold_num == fold:
            return 'Test'
        elif fold_num == (fold + 1) % 5:  # Next fold becomes validation
            return 'Valid'
        else:
            return 'Train'
    
    df_fold['Split'] = df_fold['Fold'].apply(assign_split)
    
    # Save to file
    df_fold.to_csv(f'./inputs/input_train_ptx_seg_{fold}.csv', index=False)
    
    # Print split distribution for verification
    split_counts = df_fold['Split'].value_counts()
    print(f"Fold {fold + 1} - Train: {split_counts.get('Train', 0)}, "
          f"Valid: {split_counts.get('Valid', 0)}, Test: {split_counts.get('Test', 0)}")
```
