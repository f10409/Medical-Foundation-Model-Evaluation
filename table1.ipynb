```python
import pandas as pd
```

### PTX

```python
df = pd.read_csv('./inputs/SIIM_ACR_input.csv')
df_balanced = df.groupby('Pneumothorax').sample(n=2379, random_state=5656)
df_balanced
```

```python
ptx_vol = pd.read_csv('./inputs/ptx_volume_pct.csv')
ptx_vol
```

```python
tube = pd.read_csv('/mnt/NAS3/datasets/external/SIIM_ACR_Pneumothorax/MDai_Label_SIIM.csv')
tube['ImagePath'] = tube.ImagePath.apply(lambda x: x.split('/')[-1][:-4]+'.png')
tube = tube[['ImagePath','ChestTube']]
tube
```

```python
df_balanced = df_balanced.merge(ptx_vol, on='ImagePath', how='left')
df_balanced = df_balanced.merge(tube, on='ImagePath')
df_balanced
```

```python
from tableone import TableOne
```

```python
columns = ['Pneumothorax', 'ChestTube', 'Volume', 'Volume Quartile']
categorical = ['Pneumothorax', 'ChestTube', 'Volume Quartile']
continuous = ['Volume']
groupby = 'Volume Quartile'
```

```python
mytable = TableOne(df_balanced, columns, categorical, continuous, groupby)
print(mytable.tabulate(tablefmt="github"))
```

```python
mytable.to_csv('ptx_table1.csv')
```

### Cardiomegaly

```python
df = pd.read_csv('./inputs/EmoryCXR_Cardiomegaly_input.csv')
df_balanced = df.groupby('Cardiomegaly').sample(n=2500, random_state=5656)
df_balanced
```

```python
meta = pd.read_csv('/mnt/NAS3/datasets/internal/EmoryCXRv2/TABLES/EmoryCXR_v2_Metadata_08152025.csv')
meta
```

```python
meta = meta[['empi_anon','AccessionNumber_anon','Sex','Age','Race','BMI']].drop_duplicates(subset=['empi_anon','AccessionNumber_anon'])
```

```python
df_balanced = df_balanced.merge(meta, on=['empi_anon','AccessionNumber_anon'], how='left')
```

```python
columns = ['Cardiomegaly', 'Sex', 'Age', 'BMI', 'Race']
categorical = ['Cardiomegaly', 'Sex', 'Race']
continuous = ['Age','BMI']
groupby = 'Cardiomegaly'
```

```python
mytable = TableOne(df_balanced, columns, categorical, continuous, groupby)
print(mytable.tabulate(tablefmt="github"))
```

```python
mytable.to_csv('cardiac_table1.csv')
```
